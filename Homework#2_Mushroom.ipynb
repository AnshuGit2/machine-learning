# 2.0 Necessary imports

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# 3.0 Loading data from CSV and data counts

```python
data =pd.read_csv("./agaricus-lepiota.csv", names=["class","cap-shape","cap-surface","cap-color",
                                                   "bruises?","odor","gill-attachment",
                                                   "gill-spacing","gill-size","gill-color",
                                                   "stalk-shape","stalk-root",
                                                   "stalk-surface-above-ring","stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring","veil-type","veil-color","ring-number","ring-type","spore-print-color","population","habitat"])
data
```

```python
data.head()
```

# 4.0 Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in data.columns:
    data[col] = le.fit(data[col]).transform(data[col])
```

```python
data.head()
```

```python
data.describe()
```

```python
x = data.drop(['class'], axis=1)
y = data['class']
```

```python
data.describe()
```

# 5.0 Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

y = data["class"].values

x = data.drop(["class"],axis=1)

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size = 0.25)
```

# 6.0 Decision Tree Model Build

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=7)
dt.fit(x_train,y_train)

y_train_pred = dt.predict(x_train)
print('train Accuracy: %.3f' % accuracy_score(y_train, y_train_pred))

y_test_pred = dt.predict(x_test)
print('train Accuracy: %.3f' % accuracy_score(y_test, y_test_pred))

scores = cross_val_score(dt, x_train, y_train, cv=5, scoring = "accuracy")
print('cross val Accuracy: %.3f' % np.mean(scores))

```

# 7.0 Feature Importance

```python
import graphviz
```

```python
from sklearn import tree
```

```python
tree_clf = DecisionTreeClassifier(max_depth=7, random_state=42)
tree_clf.fit(x_train, y_train)
bars = ["cap-shape","cap-surface","cap-color","bruises?","odor","gill-attachment",
                                                   "gill-spacing","gill-size","gill-color",
                                                   "stalk-shape","stalk-root",
                                                   "stalk-surface-above-ring","stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring","veil-type","veil-color","ring-number","ring-type","spore-print-color","population","habitat"]
x_pos = np.arange(len(bars))
importance = tree_clf.feature_importances_



for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.7f' % (i,v))
fig = plt.figure(figsize=(32, 6))
plt.bar([x for x in range(len(importance))], importance)
plt.title('IRIS Decision Tree Classifier - Feature Importance')
plt.xlabel('X variables')
plt.ylabel('Importance')
plt.xticks(x_pos, bars)
plt.show()
```
